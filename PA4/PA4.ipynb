````python
## Pegasos Code
import numpy as np

def pegasos(X, Y, lambda_, iterations=1000):
    t = 0
    w = np.zeros(X.shape[1])

    for _ in range(iterations):

        # select random training example
        idx = np.random.randint(0, X.shape[0])
        x = X[idx]
        y = Y[idx]

        t += 1

        # learning rate
        eta = 1 / (lambda_ * t)

        # compute score
        score = np.dot(w, x)

        # update rule
        if y * score < 1:
            w = (1 - eta * lambda_) * w + eta * y * x
        else:
            w = (1 - eta * lambda_) * w

    return w
        
````
